```python
#Visualisasi array (matriks 2D/3D)
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
# 1. membaca gambar asli 
img = Image.open ("bwek.png")
# 2. konversi gambar ke array numpy
img_array = np.array(img)
# 3. tampilkan gambar asli
plt.imshow(img_array)
plt.title("gambar asli")
plt.axis("off")
plt.show()
# 4. visualisasi channel RGB dengan heatmap
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
channels = ['Red', 'Green', 'Blue']
for i, ax in enumerate(axes):
    ax.imshow(img_array[:, :, i], cmap='inferno')
    ax.set_title(f"Channel {channels[i]}")
    ax.axis("off")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

plt.suptitle("Heatmap Array Numpy dari Channel RGB", fontsize=14)
plt.show()
#5. Koversi ke Grayscale
gray = 0.299*img_array[:,:,0] + 0.587*img_array[:,:,1] + 0.114*img_array[:,:,2]
plt.imshow(gray, cmap='gray')
plt.title("Grayscale (NumPy Array)")
plt.axis("off")
plt.show()  
#6.Edge detection sederhana
diff_x = np.abs(np.diff(gray, axis=1))  # tepi horizontal
diff_y = np.abs(np.diff(gray, axis=0))  # tepi vertikal
# pad agar ukuran sama dengan gray
diff_x = np.pad(diff_x, ((0, 0), (0, 1)), mode='constant') # tambahkan kolom kosong
diff_y = np.pad(diff_y, ((0, 1), (0, 0)), mode='constant')  # tambahkan baris kosong

edges = diff_x + diff_y

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (NumPy)")
plt.axis("off")
plt.show()
```

```python
#7. Manipulasi array (flip & rotate)
flipped = np.flipud(img_array)  # flip vertical
rotated = np.rot90(img_array)   # rotasi 90 derajat

fig, axes = plt.subplots(1, 2, figssize=(10,5))
axes[0].imshow(flipped)
axes[0].set_title("Flip Vertikal")
axes[0].axis("off")

axes[1].imshow(rotated)
axes[1].set_title("Rotasi 90°")
axes[1].axis("off")

plt.suptitle("Manipulatif Citra dengan NumPy", fontsize=14)
plt.show()
```

```python
# 13.1 numpy.amin() dan numpy.amax()
arr = np.array([3,1,4,7,4,3,2,1,9])
min_value = np.amin(arr)
max_value = np.amax(arr)
print("Array:", arr)
print("Nilai minimum:", min_value)
print("Nilai maksimum:", max_value)
```

```python
#13.2 numpy.ptp()
arr = np.array([3,1,4,7,4,3,2,1,9])
range_value = np.ptp(arr)
print("Rentang nilai:", range_value)

#13.3 numpy.percentile()
arr = np.array([3,1,4,7,4,3,2,1,9])
q1 = np.percentile(arr, 25)
q2 = np.percentile(arr, 50)
q3 = np.percentile(arr, 75)
print("Kuartil 1:", q1)
print("Kuartil 2:", q2)
print("Kuartil 3:", q3)
```

```python
#13.4 numpy median()
arr = np.array([3,1,4,7,4,3,2,1,9])
median_value = np.median(arr)
print("Median:", median_value)
```

```python
#13.5 numpy.mean()
arr = np.array([3,1,4,7,4,3,2,1,9])
mean_value = np.mean(arr)
print("Mean:", mean_value)
```

```python
#13.6 numpy.average()
arr = np.array([3,1,4,7,4,3,2,1,9])
average = np.average(arr)
print("Average:", average)
```

```python
#13.7 Standars Deviation
arr = np.array([3,1,4,7,4,3,2,1,9])
std_value = np.std(arr)
print("Standar Deviasi:", std_value)
```

```python
#13.8 Variance
arr = np.array([3,1,4,7,4,3,2,1,9])
variance = np.var(arr)
print("Variance:", variance)
```

```python
#13.9 Data Kelompok
import numpy as np
print("---DATA KELOMPOK----")
## Data berkelompok
interval = np.array([30,40,50,60,70,80])
frekuensi = np.array([8, 15, 21, 13, 10])
print("Interval:", interval)
print("Frekuensi:", frekuensi)
## Menghitung midpoint dari setiap interval
midpoint = (interval[:-1] + interval[1:]) / 2
print("Midpoint:", midpoint)
## Menghitung mean dari data berkelompok
mean = np.sum(midpoint * frekuensi) / np.sum(frekuensi)
print("Mean:", mean)
## Menghitung variance dari data berkelompok
variance = np.sum((midpoint - mean)**2 * frekuensi) / np.sum(frekuensi)
print("Variance:", variance)
## Menghitung standar deviasi dari data berkelompok
std_deviation = np.sqrt(variance)
print("Standar Deviasi:", std_deviation)
```

```python
import numpy as np
#Operasi Statistika
data = np.array([6,11,18,27,38])
mean = np.mean(data)
median = np.median(data)
std = np.std(data)
variance= np.var(data)
min_value = np.amin(data)
max_value = np.amax(data)
print("Mean:", mean)
print("Median:", median)
print("Standar Deviasi:", std)
print("Variance:", variance)
print("Nilai minimum:", min_value)
print("Nilai maksimum:", max_value)

#Menghitung korelasi antara dua set data
data1 = np.array([13,23,60,43,33])
data2 = np.array([6,7,8,9,10])
correlation = np.corrcoef(data1, data2)
print("Korelasi antara data1 dan data2:",correlation)
```
